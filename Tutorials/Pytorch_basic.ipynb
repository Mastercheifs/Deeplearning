딥러닝을 가능케한 autograd  (쉽게 이야기하면 미분하게 해주는 거)

```python
x = torch.tensor([1.], requires_grad=True) #float여야해서 1.
print(x)
```

```python
x = torch.tensor([1.])
print(x)
print(x.requires_grad)

x.requires_grad = True
print(x)
print(x.requires_grad)
```

```python
x = torch.tensor([1.], requires_grad=True)
y = x**2
print(y)

print(x.grad)
y.backward() #미분을 하는 함수다
print(x.grad) #y = x**2을 미분한 2x의 x 값에 1을 대입한 gradient 값
```

```python
x = torch.tensor([1.], requires_grad=True)
y = x**2
print(y)

z = 3*y
print(z)

z.backward() #미분을 하는 함수다
print(x.grad)
```

```python
x = torch.tensor([1.], requires_grad=True)
y = x**2
print(y)
y.retain_grad() # 이걸 하면 y.grad도 볼 수 있다

z = 3*y
print(z)

z.backward() #미분을 하는 함수다
print(y.grad) 
```

```python
#detach의 사용 용도
x = torch.tensor([1.], requires_grad=True)
y = x**2 # y = x^2
z = y.detach() #x로 만든 것을 상수로 사용하고 싶은 것. z 는 1이 된다
w = y+z # x**2 + 1

w.backward()
print(x.grad)
```

```python
# 많이 쓰이는 torch.no_grad
x = torch.tensor([1.], requires_grad=True)
# chain rule을 위해 계속 grad_fn 을 update 하니까 grad_fn 잠시 안계산하고 싶을때 torch.no.grad
# 모델 테스트 시에는 불필요하게 메모리를 쓸 필요가 없기 때문

with torch.no_grad():
    y = x**2
    print(x.requires_grad)
    print(y)
print(x.requires_grad)

y = x**2
print(y)

x = torch.tensor([1.], requires_grad = True)
x. requires_grad = False
y = x**2
print(x.requires_grad)
print(y)

```

```python
class Mymodel(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5)
        self.fc2 = nn.Linear(5, 10)
        self.fc3 = nn.Linear(10, 3)
        self.act = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.act(x)
        x = self.fc3(x)
        x = self.act(x)
        return x
    
model = Mymodel()
x = torch.randn(5, 2)
y = model(x)
print(y)

make_dot(y)
```
